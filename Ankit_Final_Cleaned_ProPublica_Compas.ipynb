### SVC

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from aif360.datasets import BinaryLabelDataset
from aif360.metrics import ClassificationMetric
from aif360.algorithms.postprocessing import RejectOptionClassification, CalibratedEqOddsPostprocessing, EqOddsPostprocessing

X_train, X_test, y_train, y_test = train_test_split(X, y['is_recid'], test_size=0.2, random_state=42)

clf = SVC(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

privileged_groups = [{'sex_Male': 1}]  # Example: Males are privileged
unprivileged_groups = [{'sex_Male': 0}]  # Example: Females are unprivileged

df_test = X_test.copy()
df_test['label'] = y_test
dataset = BinaryLabelDataset(df=pd.DataFrame(df_test), label_names=['label'], protected_attribute_names=['sex_Male'])

df_pred = X_test.copy()
df_pred['label'] = y_pred
dataset_pred = BinaryLabelDataset(df=pd.DataFrame(df_pred), label_names=['label'], protected_attribute_names=['sex_Male'])

metric = ClassificationMetric(dataset, dataset_pred, privileged_groups=privileged_groups, unprivileged_groups=unprivileged_groups)
print("Initial Disparate Impact:", metric.disparate_impact())

# 1. Reject Option Classification
roc = RejectOptionClassification(
    unprivileged_groups=unprivileged_groups,
    privileged_groups=privileged_groups,
    low_class_thresh=0.3,  # Adjust these
    high_class_thresh=0.7,
    num_class_thresh=100
)
roc.fit(dataset, dataset_pred)
dataset_transf_pred = roc.predict(dataset_pred)
y_pred_roc = dataset_transf_pred.labels

# 2. Calibrated Equalized Odds Post-processing
ceo = CalibratedEqOddsPostprocessing(unprivileged_groups, privileged_groups, seed=42)
ceo.fit(dataset, dataset_pred)
dataset_ceo_pred = ceo.predict(dataset_pred)
y_pred_ceo = dataset_ceo_pred.labels

# 3. Equalized Odds Post-processing
eqo = EqOddsPostprocessing(unprivileged_groups, privileged_groups)
eqo.fit(dataset, dataset_pred)
dataset_eqo_pred = eqo.predict(dataset_pred)
y_pred_eqo = dataset_eqo_pred.labels

print("Post-Processed Disparate Impact (ROC):", ClassificationMetric(dataset, dataset_transf_pred, privileged_groups, unprivileged_groups).disparate_impact())
print("Post-Processed Disparate Impact (CEO):", ClassificationMetric(dataset, dataset_ceo_pred, privileged_groups, unprivileged_groups).disparate_impact())
print("Post-Processed Disparate Impact (Equalized Odds):", ClassificationMetric(dataset, dataset_eqo_pred, privileged_groups, unprivileged_groups).disparate_impact())
```

```python
from sklearn.metrics import accuracy_score

print("Original Model Accuracy:", accuracy_score(y_test, y_pred))
print("Post-Processed Accuracy (ROC):", accuracy_score(y_test, y_pred_roc))
print("Post-Processed Accuracy (CEO):", accuracy_score(y_test, y_pred_ceo))
print("Post-Processed Accuracy (EO):", accuracy_score(y_test, y_pred_eqo))
```
